```python
sentiment_map = {
    'Extreme Fear': -2,
    'Fear': -1,
    'Neutral': 0,
    'Greed': 1,
    'Extreme Greed': 2
}
fgi['SentimentScore'] = fgi['classification'].map(sentiment_map)
```

```python
hd['Timestamp IST'] = pd.to_datetime(hd['Timestamp IST'], errors='coerce', dayfirst=True)
```

```python
hd['NetPnL'] = hd['Closed PnL'] - hd['Fee']
hd['Timestamp IST'] = pd.to_datetime(hd['Timestamp IST'])
hd['Leverage'] = hd['Size USD'] / (hd['Start Position'] * hd['Execution Price'])
```

```python
hd['Date'] = hd['Timestamp IST'].dt.date
fgi['Date'] = pd.to_datetime(fgi['date']).dt.date
```

```python
df = hd.merge(fgi[['Date','classification']],on='Date',how='left')
```

```python
df['NetPnL'] = hd['Closed PnL'] - hd['Fee']
df['SentimentScore'] = fgi['classification'].map({
    'Extreme Fear': -2,
    'Fear': -1,
    'Neutral': 0,
    'Greed': 1,
    'Extreme Greed': 2
})
df['Leverage'] = df['Size USD'] / (hd['Start Position'] * hd['Execution Price'])
```

```python
df.isnull().sum()
```

```python
df['SentimentScore'] = df['classification'].map(sentiment_map)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['SentimentScore','classification'])
```

```python
df.isnull().sum()
```

```python
df.groupby('classification')['NetPnL'].describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

avg_pnl = df.groupby('classification')['NetPnL'].mean().reindex([
    'Extreme Fear', 'Fear', 'Neutral', 'Greed', 'Extreme Greed'
])
avg_pnl.plot(kind='bar', title='Average NetPnL by Market Sentiment')
plt.ylabel('Net Profit/Loss')
plt.show()
```

```python
sns.boxplot(data=df, x='classification', y='Leverage', order=[
    'Extreme Fear', 'Fear', 'Neutral', 'Greed', 'Extreme Greed'
])
plt.ylim(0, 10)  # Optional: restrict to reasonable leverage
plt.title('Leverage Distribution by Sentiment')
plt.show()
```

```python
df['Profitable'] = df['NetPnL'] > 0
win_rate = df.groupby('classification')['Profitable'].mean()
print(win_rate)
```

# Key Insights & Patterns

### Extreme Greed shows the highest average NetPnL

- Mean PnL is $67.22, significantly higher than other sentiments.
- Also has a relatively low standard deviation, suggesting more stable positive outcomes.

 Implication: Traders perform best in Extreme Greed  ideal time to scale trades.

### Fear has surprisingly high avg. NetPnL

- With $52.79, Fear outperforms even Greed ($41.48).

Suggests that experienced traders capitalize on market fear, possibly via shorting or quick bounces.
 Implication: There are strong opportunities during Fear  but likely for advanced traders using specific strategies.

### Extreme Fear and Neutral yield similar avg. PnL (~$33)
Despite emotional extremes, performance is modest.

High standard deviation and large loss potential in Extreme Fear.
 Implication: Traders may act cautiously or exit markets during Extreme Fear, reducing potential for profit or loss.

### Greed has lowest avg. PnL after Neutral
Surprisingly, Greed ($41.48) performs worse than Fear or Extreme Greed.

But highest loss: -118,071! This suggests:

Risky trades

Overconfidence
 Implication: Greed invites high-risk behavior. Warn traders about leverage misuse here.

### Risk/Reward Profile (25th vs. 75th Percentile)
75% PnL is always positive, suggesting most trades are profitable.

25% is negative, but not extremely  losses are small for majority of traders.
 Implication: Most trades are low-risk, but a few highly negative trades skew the mean.
